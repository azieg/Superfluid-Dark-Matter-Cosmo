```python
import math
import numpy as np
import matplotlib.pyplot as plt

def distrib(ϵ, ζ):
    """Calculates the occupation number f for a chosen fugacity ζ
    over a range of energy values ϵ (mathematically ϵ/T)"""
    return 1.0 / (ζ**-1 * math.exp(ϵ)- 1.0)

n   = 100                                                   # Number of points for each energy
ϵT  = np.linspace(.009, 2, n, endpoint=True)                        # Array of energy values
ϵT2 = np.linspace(0, 2, n, endpoint=True)                                  # Used for plotting because ϵT does weird stuff
ζ   = [1.0, .5, .33, .25, .1]                               # Fugacity values
f   = [[0 for i in range(len(ϵT) + 1)] for j in range(5)]   # Empty occupation number matrix for the 5 fuacities

# Calculates the occupation numbers over
#  an energy range for each fuacity
for j in range(len(ζ)):
    for i in range(n):
        f[j][i] = distrib(ϵT[i], ζ[j])
                
# Plot label colors
clr = ['b', 'r', 'g', 'y', 'k']
ls  = ['-.', '--', ':', '-']
plt.figure(figsize = (14, 10))
lb = 'ζ = ' + str('{:04.2f}'.format(1.0))
plt.plot(ϵT2, f[0][0:-1], color = clr[0], label = lb)

for i in range(1, 5):
    lb = 'ζ = ' + str('{:04.2f}'.format(ζ[i]))
    plt.plot(ϵT2, f[i][0:-1], ls = ls[i - 1], color = clr[i], label = lb)
    
plt.xlabel('Energy ϵ/T', fontsize=20)
plt.ylabel('Occupation Number $f^0$', fontsize=20)
plt.legend(loc='best', prop={'size': 20})
plt.tick_params(labelsize=20)
plt.xlim(0, 2)
plt.ylim(0, 4)
#plt.grid()
plt.savefig("Bose_Distrib", dpi=200)
plt.show()
```
