```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 20.0
```

```python
def boseDistrib(ϵ, ζ):
    """Calculates the occupation number f for a chosen fugacity ζ
    over a range of energy values ϵ (mathematically ϵ/T)"""
    return 1.0/(ζ**-1.0*math.exp(ϵ) - 1.0)
```

```python
ENERGIES_N = 100
FUGACITY_N = 5

energies = np.linspace(.009, 2, ENERGIES_N, endpoint=True)
fugacity = np.linspace(1.0, .1, FUGACITY_N, endpoint=True)

occupation_num = np.ndarray((FUGACITY_N, ENERGIES_N,), dtype = np.longdouble)

for i in range(ENERGIES_N):
    for j in range(FUGACITY_N):
        occupation_num[j][i] = boseDistrib(energies[i], fugacity[j])

clr = ['b', 'r', 'g', 'y', 'k']
ls  = ['-', '-.', '--', ':', '-']
```

```python
plt.figure(figsize=(14, 10))

lb = 'ζ = ' + str('{:04.2f}'.format(1.0))

for i in range(FUGACITY_N):
    lb = 'ζ = ' + str('{:04.2f}'.format(fugacity[i]))
    plt.plot(energies, occupation_num[i], ls=ls[i], color=clr[i], label=lb)
    
plt.xlabel('Energy ϵ/T')
plt.ylabel('Occupation Number $f^0$')
plt.legend(loc='best', prop={'size': 20})
plt.xlim(0, 2)
plt.ylim(0, 4)
#plt.savefig("Bose_Distrib", dpi=200)
plt.show()
```
