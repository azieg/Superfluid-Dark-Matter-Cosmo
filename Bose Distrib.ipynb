```python
import math
import numpy as np
import matplotlib.pyplot as plt

def distrib(, ):
    """Calculates the occupation number f for a chosen fugacity 
    over a range of energy values  (mathematically /T)"""
    return 1.0 / (**-1 * math.exp()- 1.0)

n   = 100                                                   # Number of points for each energy
T  = np.linspace(.009, 2, n, endpoint=True)                        # Array of energy values
T2 = np.linspace(0, 2, n, endpoint=True)                                  # Used for plotting because T does weird stuff
   = [1.0, .5, .33, .25, .1]                               # Fugacity values
f   = [[0 for i in range(len(T) + 1)] for j in range(5)]   # Empty occupation number matrix for the 5 fuacities

# Calculates the occupation numbers over
#  an energy range for each fuacity
for j in range(len()):
    for i in range(n):
        f[j][i] = distrib(T[i], [j])
                
# Plot label colors
clr = ['b', 'r', 'g', 'y', 'k']
ls  = ['-.', '--', ':', '-']
plt.figure(figsize = (14, 10))
lb = ' = ' + str('{:04.2f}'.format(1.0))
plt.plot(T2, f[0][0:-1], color = clr[0], label = lb)

for i in range(1, 5):
    lb = ' = ' + str('{:04.2f}'.format([i]))
    plt.plot(T2, f[i][0:-1], ls = ls[i - 1], color = clr[i], label = lb)
    
plt.xlabel('Energy /T', fontsize=20)
plt.ylabel('Occupation Number $f^0$', fontsize=20)
plt.legend(loc='best', prop={'size': 20})
plt.tick_params(labelsize=20)
plt.xlim(0, 2)
plt.ylim(0, 4)
#plt.grid()
plt.savefig("Bose_Distrib", dpi=200)
plt.show()
```
